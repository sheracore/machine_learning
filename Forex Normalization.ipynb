# Multiple linear regression on forex dataset

```python
import pandas as pd
import numpy asnp
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/forex_datasets/Foreign_Exchange_Rates.csv")
print(df.columns)
df.head()
```

```python
cdf = df[["AUSTRALIA - AUSTRALIAN DOLLAR/US$","EURO AREA - EURO/US$","NEW ZEALAND - NEW ZELAND DOLLAR/US$","UNITED KINGDOM - UNITED KINGDOM POUND/US$","CHINA - YUAN/US$","CANADA - CANADIAN DOLLAR/US$"]]
```

```python
cdf.head()
```

```python
plt.scatter(cdf["CANADA - CANADIAN DOLLAR/US$"],cdf["EURO AREA - EURO/US$"], color='blue')
plt.xlabel("AUSTRALIAN DOLLAR")
plt.ylabel("EURO/US")
plt.show()
```

```python
fig = plt.figure()
pic = fig.add_subplot(111)
pic.scatter(cdf["CANADA - CANADIAN DOLLAR/US$"],cdf["EURO AREA - EURO/US$"], color='blue')
pic.scatter(cdf["CANADA - CANADIAN DOLLAR/US$"],cdf["CHINA - YUAN/US$"], color='red')
plt.xlabel("CANADA - CANADIAN DOLLAR/US$")
plt.ylabel("EURO-AREA-EURO/US$:blue CHINA-YUAN/US$:red")
```
